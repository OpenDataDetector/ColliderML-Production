```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import uproot
from tqdm import tqdm
```

```python
base_dir = "/global/cfs/projectdirs/m4958/data/ColliderML/outputs/low_pileup_pilot/chichi/v2/runs"
```

```python
sample = uproot.open(os.path.join(base_dir, "0/tracksummary_ambi.root"))
sample.keys()

print('tracksummary;1' in sample.keys())
```

```python
# Loop over all run directories
run_dirs = [os.path.join(base_dir, d) for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

bad_runs = []

# Loop over all run directories
for i, run_dir in tqdm(enumerate(run_dirs)):
    # uproot load tracksummary_ambi.root in each run directory
    tracksummary_file = os.path.join(run_dir, "tracksummary_ambi.root")
    tracksummary = uproot.open(tracksummary_file)
    
    # Find runs WITHOUT tracksummary key in root file
    if "tracksummary;1" not in tracksummary.keys():
        # print(f"Run {run_dir} does not have tracksummary key")
        bad_runs.append(run_dir)

print(bad_runs)
```
